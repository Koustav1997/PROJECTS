```python
import gzip
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)
        
def getdf(path):
    i = 0
    df1 = {}
    for d in parse(path):
        df1[i] = d
        i = i+1
    return pd.DataFrame.from_dict(df1, orient='index')

  
```

```python
import re

def cleaning(text):
    text=re.sub(r'[,!*£()+.?]',' ',text) 
    text=''.join([i for i in text if not i.isdigit()])## used to remove the numerical values
    text=' '.join(text.split()) ## use to remove the extra space
    text=re.sub(r'[,.+£"#;:''@=//]','',text) ## use to remove the special characters
    text=re.sub('-(?!\w)','',text) ## removes hyphon
    text=re.sub(r'[^\w\s]','',text) ## removes the punctuations 
    text=re.sub(r"\\b(\\w+)(?:\\W+\\1\\b)+",'',text) #removes the repeted words
    text=text.lower()

    return text
```

```python
home_review_df['reviewText']=home_review_df['reviewText'].apply(cleaning)
home_review_df['summary']=home_review_df['summary'].apply(cleaning)
```

```python
home_review_df['reviewText']=home_review_df['reviewText'].astype(str)
```

```python
## CONCATENATE REVIEWTEXT AND SUMMARY 

home_review_df['review_text'] = home_review_df[['summary', 'reviewText']].apply(lambda x: " ".join(str(y) for y in x if str(y) != 'nan'), axis = 1)
home_review_df1 = home_review_df.drop(['reviewText', 'summary'], axis = 1)
home_review_df1.head()
```

```python
# Split 'helpful' column into 'neg_feedback' and 'pos_feedback' and drop it at the end
list_positive = []
list_negative = []

for help in home_review_df1['helpful']:
    list_positive.append(help[0])
    list_negative.append(help[1] - help[0])
    
# Positive Feedback (Supporting reviewer's idea and find it useful)
home_review_df1['pos_feedback'] = list_positive

# Negative Feedback (Finding reviewer's idea useless and not supporting it)
home_review_df1['neg_feedback'] = list_negative

# Drop redundant 'helpful' column
home_review_df1 = home_review_df1.drop('helpful', axis = 1)
home_review_df1.head()
```

```python
# Classify ratings as good
good_rate = len(home_review_df1[home_review_df1['overall'] >= 3])
bad_rate = len(home_review_df1[home_review_df1['overall'] < 3])

# Printing rates and their total numbers
print ('Good ratings : {} '.format(good_rate))
print ('Bad ratings : {}'.format(bad_rate))

# Apply the new classification to the ratings column
home_review_df1['rating_class'] = home_review_df1['overall'].apply(lambda x: 'bad' if x < 3 else'good')
home_review_df1.head()
```

```python
# Convert time object to datetime and create a new column named 'time'
home_review_df1['Time'] = home_review_df1.reviewTime.str.replace(',', "")
home_review_df1['Time'] = pd.to_datetime(home_review_df1['Time'], format = '%m %d %Y')

# Drop redundant 'reviewTime' column
home_review_df1 = home_review_df1.drop('reviewTime', axis = 1)
```

```python
def tokenize(column):
    #Tokenizes a Pandas dataframe column and returns a list of tokens.
    tokens = nltk.word_tokenize(column)
    return [w for w in tokens if w.islower()]  

home_df1['tokenized'] = home_df1['review_text'].apply(tokenize)
home_df1[['tokenized']].head()
```

```python
import nltk
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
home_df1['tokenized'] = home_df1['tokenized'].apply(lambda w: [w for w in w if w not in stop_words])
```

```python
from textblob import TextBlob
# function to calculate polarity
def getPolarity(review):
    return TextBlob(review).sentiment.polarity

# Polarity – talks about how positive or negative the opinion is
## Polarity ranges from -1 to 1 (1 is more positive, 0 is neutral, -1 is more negative)

# function to analyze the reviews
def analysis(score):
    if score < 0:
        return 0 #negative
    elif score == 0:
        return 2 #neutral
    else:
        return 1 #positive
```

```python
home_df1['tokenized'] = home_df1['tokenized'].astype(str)
```

```python
home_df1['Polarity'] = home_df1['tokenized'].apply(getPolarity) 
home_df1[['Polarity']]
```

```python
def sentiment(score):
    if score < 0:
        return 'Negative' 
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
home_df1['Sentiment'] = home_df1['Polarity'].apply(analysis)
home_df1['Sentiment'].value_counts()
```

```python
home_df1['Sentiment_Analysis'] = home_df1['Polarity'].apply(sentiment)
home_df1['Sentiment_Analysis'].value_counts()
```

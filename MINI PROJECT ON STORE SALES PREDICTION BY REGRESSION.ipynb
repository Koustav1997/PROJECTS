```python
x=df_scaled[['Outlet_Type','Item_MRP','Item_Visibility','Outlet_Identifier']].values.reshape(-1,4)
z=df_scaled['Item_Outlet_Sales']
y=np.array(z)
```

```python
from sklearn.linear_model import LinearRegression
#importing linear regreesion model
```

```python
model=LinearRegression()
#defining the model
```

```python
y_pred=model.predict(x_test)
y_pred
#predicting data using the model
```

```python
model.score(x_train,y_train)
#model accuacy on train data
```

```python
model.score(x_test,y_test)
#model accuracy on test data
```

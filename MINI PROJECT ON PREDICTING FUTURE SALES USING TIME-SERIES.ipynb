# Checking Stationarity of data

```python
from statsmodels.tsa.stattools import adfuller,acf,pacf
def checkstationarity(data):
    pvalue=adfuller(data)[1]
    print(pvalue)
    if pvalue < 0.05:
        ret='Data is stationary go for model building'
    else:
        ret='Data is not stationary make it stationary before model building'
    
    return ret
```

```python
checkstationarity(data_w)
```

```python
#from augmented-dickey-fullers test(ADF) and also from the linear downward trend which we observed 
#we can confirm that the data is stationary
```
